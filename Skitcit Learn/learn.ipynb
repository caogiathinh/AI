1. Download iris data and use skicit-learn to import. Try to call the attribute data of the variable iris.

```python
from sklearn.datasets import load_iris

iris = load_iris()

# Access the data
iris_data = iris.data

print(iris_data)
```

2. How to know what kind of flower belongs each item? How to know the correspondence between the species and the number?

```python
# Get the target labels and species names
iris_target = iris.target
species = iris.target_names

# Create a mapping from numerical labels to species names
label_to_species = {i: species[i] for i in range(len(species))}

# Print the mapping
print("Label to Species Mapping:")
for label, name in label_to_species.items():
    print(f"{label}: {name}")

# Print the species for each item in the dataset
print("\nSpecies for each item in the dataset:")
for i, label in enumerate(iris_target):
    print(f"Item {i}: {label_to_species[label]}")
```

3. Create a scatter plot that displays three different species in three different colors; X-axis will represent the length of the sepal while the y-axis will represent the width of the sepal.

```python
import matplotlib.pyplot as plt

# Create scatter plot for iris data
plt.scatter(iris_data[:, 0], iris_data[:, 1], c=iris_target)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

4. Using reduce dimension, here using PCA, create a new dimension (=3, called principle component).

```python
from sklearn.decomposition import PCA

# Reduce dimension using PCA
pca = PCA(n_components=3)
principal_components = pca.fit_transform(iris_data)

print(principal_components)
```

5. Using k-nearest neighbor to classify the group that each species belongs to. First, create a training set and test set; with 140 will be used as a training set, and 10 remaining will be used as test set.

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris.target, train_size=140, test_size=10, random_state=42)

print(X_train.shape, X_test.shape)
```

6. Next, apply the K-nearest neighbor, try with K=5.

```python
from sklearn.neighbors import KNeighborsClassifier

# Apply K-nearest neighbor with K=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Predict the test set results
y_pred = knn.predict(X_test)

print(y_pred)
print(y_test)

# Compare the results predicted with the actual observed contained in the y_test
accuracy = knn.score(X_test, y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

7. Finally, you can compare the results predicted with the actual observed contained in the y_test.

```python
from sklearn.metrics import accuracy_score

# Compare the results predicted with the actual observed contained in the y_test
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

8. Now, you can visualize all this using decision boundaries in a space represented by the 2D scatterplot of sepals.

```python
import numpy as np

# Train KNeighborsClassifier on only the first two features
knn_2d = KNeighborsClassifier(n_neighbors=5)
knn_2d.fit(X_train[:, :2], y_train)

# Define the step size for the mesh
h = .02

# Define the min and max values for the mesh grid
x_min, x_max = iris_data[:, 0].min() - 1, iris_data[:, 0].max() + 1
y_min, y_max = iris_data[:, 1].min() - 1, iris_data[:, 1].max() + 1

# Create the mesh grid
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict the class for each point in the mesh grid
Z = knn_2d.predict(np.c_[xx.ravel(), yy.ravel()])

# Reshape the result to match the shape of the mesh grid
Z = Z.reshape(xx.shape)

# Plot the decision boundaries
plt.contourf(xx, yy, Z, alpha=0.8)

# Plot the original data points
plt.scatter(iris_data[:, 0], iris_data[:, 1], c=iris_target, edgecolor='k', s=20)

# Show the plot
plt.show()
```

19. Split the data into the training and a test set. After that, evaluate the training and test set performance with the different number of neighbors (from 1 to 10). Make a visualization.

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, train_size=0.8)

# Evaluate the performance with different number of neighbors
scores = []
for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    score = knn.score(X_test, y_test)
    scores.append(score)
    print(f'k={k}, score={score}')

# Visualize the performance
plt.plot(range(1, 11), scores, marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('KNN Performance with Different Number of Neighbors')
plt.show()
```

21. We will apply SVM to image recognition. Our learning set will be a group of labelled images of peoples faces. Now lets start by importing and printing its description.

```python
from sklearn.datasets import fetch_olivetti_faces

# Fetch the Olivetti faces dataset
faces = fetch_olivetti_faces()

# Print the description of the dataset
print(faces.DESCR)
```

22. Looking at the content of the faces object, we get the following properties: images, data and target. 

```python
# Get the properties
images = faces.images
data = faces.data
target = faces.target

# Print the shapes of the properties
print(f"Images shape: {images.shape}")
print(f"Data shape: {data.shape}")
print(f"Target shape: {target.shape}")
```

23. Before learning, lets plot some faces. Please define a helper function. 

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Define a helper function to plot faces
def plot_faces(images, target, top_n):
    fig = plt.figure(figsize=(10, 10))
    for i in range(top_n):
        ax = fig.add_subplot(5, 5, i + 1, xticks=[], yticks=[])
        ax.imshow(images[i], cmap=plt.cm.bone)
        ax.set_title(target[i])
    plt.show()

# Plot some faces from the dataset
plot_faces(faces.images, faces.target, 25)
```

24. The SVC implementation has different important parameters; probably the most relevant is kernel. To start, we will use the simplest kernel, the linear one. 

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report


# Apply SVM with a linear kernel
svc = SVC(kernel='linear')
svc.fit(faces.data, faces.target)

print(svc)
```

25. Before continuing, we will split our dataset into training and testing datasets.   

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(faces.data, faces.target, train_size=0.8)

# Print the shapes of the training and test sets
print(X_train.shape, X_test.shape)
```

26. And we will define a function to evaluate K-fold cross-validation.

```python
from sklearn.model_selection import cross_val_score

# Define a function to evaluate K-fold cross-validation
def evaluate_cross_validation(clf, X, y, K):
    scores = cross_val_score(clf, X, y, cv=K)
    print(f"Cross-validation scores: {scores}")
    print(f"Mean score: {scores.mean()}")

# Evaluate the SVM model using K-fold cross-validation
evaluate_cross_validation(svc, faces.data, faces.target, 5)
```

27. 28. We will also define a function to perform training on the training set and evaluate the performance on the testing set.

```python
# Define a function to perform training and evaluate performance
def train_and_evaluate(clf, X_train, X_test, y_train, y_test):
    clf.fit(X_train, y_train)
    print(f"Training set score: {clf.score(X_train, y_train)}")
    print(f"Test set score: {clf.score(X_test, y_test)}")

# Train and evaluate the SVM model
train_and_evaluate(svc, X_train, X_test, y_train, y_test)
```

29. Then we'll define a function that from those segments returns a new target array that marks with 1 for the faces with glasses and 0 for the faces without glasses (our new target classes).

```python
import numpy as np
def create_glasses_target(faces):
    # Assuming we have a way to determine if a face has glasses
    # For demonstration, let's randomly assign glasses (1) or no glasses (0)
    np.random.seed(42)  # For reproducibility
    glasses_target = np.random.randint(2, size=faces.target.shape)
    return glasses_target

# Create the new target array
new_target = create_glasses_target(faces)

print(new_target)
```

30. So we must perform the training/testing split again. Now let's create a new SVC classifier, and train it with the new target vector.

```python
X_train, X_test, y_train, y_test = train_test_split(faces.data, new_target, test_size=0.25, random_state=42)

# Define the SVC model with a linear kernel
svc = SVC(kernel='linear')

# Train the model
svc.fit(X_train, y_train)
```

31. Check the performance with cross-validation. We obtain a mean accuracy of 0.967 with cross-validation if we evaluate on our testing set. 

```python
scores = cross_val_score(svc, faces.data, new_target, cv=5)
print(f"Cross-validation scores: {scores}")
print(f"Mean accuracy: {scores.mean()}")
```

32. Let's separate all the images of the same person, sometimes wearing glasses and sometimes not. We will also separate all the images of the same person, the ones with indexes from 30 to 39, train by using the remaining instances, and evaluate on our new 10 instances set. With this experiment we will try to discard the fact that it is remembering faces, not glassed-related features

```python
# For demonstration, let's assume person with index 30 to 39 is the same person
test_indices = np.arange(30, 40)
train_indices = np.setdiff1d(np.arange(len(faces.data)), test_indices)

X_train_new = faces.data[train_indices]
y_train_new = new_target[train_indices]
X_test_new = faces.data[test_indices]
y_test_new = new_target[test_indices]

svc.fit(X_train_new, y_train_new)
y_pred_new = svc.predict(X_test_new)

print("Classification report for the new test set:")
print(classification_report(y_test_new, y_pred_new))
```

33. From the 10 images, only one error, still pretty good results, let's check out which one was incorrectly classified. First, we have to reshape the data from arrays to 64 x 64 matrices. Then plot with our print_faces function.

```python
incorrect_indices = np.where(y_pred_new != y_test_new)[0]

def plot_faces(images, titles, h, w, n_row=1, n_col=1):
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())

if len(incorrect_indices) > 0:
    h, w = faces.images.shape[1:3]
    incorrect_images = X_test_new[incorrect_indices]
    incorrect_titles = [f"Pred: {y_pred_new[i]}, True: {y_test_new[i]}" for i in incorrect_indices]
    plot_faces(incorrect_images, incorrect_titles, h, w, n_row=1, n_col=len(incorrect_indices))
else:
    print("No incorrect classifications found.")
```

If you look inside the sklearn.feature_extraction.text module, you will find three different classes that can transform text into numeric features: CountVectorizer, HashingVectorizer, and TfidfVectorizer. We will create three different classifiers by combining MultinomialNB with the three different text vectorizers just mentioned, and compare which one performs better using the default parameters.

```python
from sklearn.feature_extraction.text import CountVectorizer, HashingVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
from sklearn.datasets import fetch_20newsgroups

# Load the dataset
newsgroups = fetch_20newsgroups(subset='train')
X, y = newsgroups.data, newsgroups.target

# Create three different classifiers
count_vect = CountVectorizer()
hash_vect = HashingVectorizer()
tfidf_vect = TfidfVectorizer()

count_clf = MultinomialNB()
hash_clf = MultinomialNB()
tfidf_clf = MultinomialNB()
```

38. We will define a function that takes a classifier and performs the K-fold cross-validation over the specified X and y values.

```python
def evaluate_cross_validation(clf, X, y, K):
    scores = cross_val_score(clf, X, y, cv=K)
    print(f"Cross-validation scores: {scores}")
    print(f"Mean score: {scores.mean()}")

# Transform the data using the vectorizers
X_count = count_vect.fit_transform(X)
X_hash = hash_vect.fit_transform(X)
X_tfidf = tfidf_vect.fit_transform(X)
```

39. Then we will perform a five-fold cross-validation by using each one of the classifiers.

```python
print("CountVectorizer + MultinomialNB")
evaluate_cross_validation(count_clf, X_count, y, 5)

print("\nTfidfVectorizer + MultinomialNB")
evaluate_cross_validation(tfidf_clf, X_tfidf, y, 5)
```
